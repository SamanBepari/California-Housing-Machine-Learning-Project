```python
x= california_df.drop('MedHouseVal',axis=1)  # input  , axis=0 means rows, axis =1 means columns
y= california_df['MedHouseVal']              # output
```

**Find the VIF**

```python
# class----> statsmodels.stats.outliers_influence method
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_data=[]   ## empty list  later used in a dataframe
for i in range(len(x.columns)):
    vif_data.append(variance_inflation_factor(x.values,i))   ## vifs are calculated and stored in empty list
vif_data   
```

**Create Dataframe**

```python
import pandas as pd
vif_df=pd.DataFrame(vif_data,index=x.columns,columns=['VIF'])  ## taking index as the input columns for better view
vif_df.sort_values(by='VIF',ascending=False)  ## just sorting the column VIF in descending order
```

```python
# till now we have seen that the dataframe method has index,the read datset, columns, axis as different attributes
```

**Select the Final columns**

```python
con1 = vif_df[vif_df['VIF']<15].index
con1 = con1.to_list()
con1
```

```python
b=vif_df[vif_df['VIF']>15].index
b
```

```python
con2 = ['AveBedrms','Latitude']
con2
```

```python
final_cols=con1+con2
final_cols
```

**Created a Final DataFrame (VIF)**

```python
import warnings
warnings.filterwarnings('ignore')
final_df=x[final_cols]
final_df['MedHouseVal']=y
final_df
```

```python
X=final_df.drop('MedHouseVal',axis=1)
Y=final_df['MedHouseVal']
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x_scale=ss.fit_transform(X)
x_scale
```

**Created a Scale DF**

```python
x_scale_df=pd.DataFrame(x_scale,columns=X.columns)
x_scale_df
```

**Method : Train Test Split**

- Package name : sklearn
 - class name   : model_selection
 - Method Name  : Train_Test_split

```python
from sklearn.model_selection import train_test_split
# Train the Model

x_train,x_test,y_train,y_test=train_test_split(x_scale_df,
                                              Y,
                                              test_size=0.2,
                                              random_state=1234)
```
